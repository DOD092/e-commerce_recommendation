```python
df_product_clean = df_product.drop(columns=["seller"])
```

```python
df_product_clean['price'] = df_product_clean['price'].astype('float')
```

```python
df_product_clean = df_product_clean[df_product_clean['price'] > 0]
df_product_clean = df_product_clean[df_product_clean['rating'].between(0,5)]
```

```python
df_product_clean['short_description'] = df_product_clean['short_description'].str.lower().str.replace(r'[^\w\s]', '', regex=True)
```
