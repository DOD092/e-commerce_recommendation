```python
df_cf['customer_id'] = df_cf['customer_id'].astype(int)
df_cf['product_id']  = df_cf['product_id'].astype(int)
```

```python
user_item = df_cf.pivot_table(
    index='customer_id',
    columns='product_id',
    values='rating',
    aggfunc='mean'   
)
```

```python
def top_k_neighbors(sim_series: pd.Series, k: int, exclude_self: bool = True):
    s = sim_series.copy()
    if exclude_self:
        s = s.drop(sim_series.name, errors='ignore')
    return s.sort_values(ascending=False).head(k)
```

```python
def get_unrated_items(user_item: pd.DataFrame, user_id: int):
    if user_id not in user_item.index:
        return []
    user_ratings = user_item.loc[user_id]
    return user_ratings[user_ratings.isna()].index.tolist()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
user_item_filled = user_item.fillna(0)
```

```python
user_sim = pd.DataFrame(
    cosine_similarity(user_item_filled),
    index=user_item.index,
    columns=user_item.index
)
```

```python
item_sim = pd.DataFrame(
    cosine_similarity(user_item_filled.T),
    index=user_item.columns,
    columns=user_item.columns
)
```

```python
def predict_user_based(user_id: int, item_id: int, user_item: pd.DataFrame, user_sim: pd.DataFrame, k: int = 20):
    # Check tồn tại
    if user_id not in user_item.index or item_id not in user_item.columns:
        return np.nan

    # Các user đã rating item_id
    raters = user_item[item_id].dropna()
    if raters.empty:
        return np.nan

    # Similarity giữa user_id và các user đã rating item đó
    sims = user_sim.loc[user_id, raters.index]
    # Lấy top-k neighbors
    top = sims.sort_values(ascending=False).head(k)

    # Weighted average
    ratings = raters.loc[top.index]
    denom = np.sum(np.abs(top.values))
    if denom == 0:
        return np.nan
    return float(np.dot(top.values, ratings.values) / denom)
```

```python
def predict_item_based(user_id: int, item_id: int, user_item: pd.DataFrame, item_sim: pd.DataFrame, k: int = 20):
    if user_id not in user_item.index or item_id not in user_item.columns:
        return np.nan

    # Các item user đã rating
    user_ratings = user_item.loc[user_id].dropna()
    if user_ratings.empty:
        return np.nan

    # Similarity giữa item_id và các item user đã rating
    sims = item_sim.loc[item_id, user_ratings.index]
    top = sims.sort_values(ascending=False).head(k)

    ratings = user_ratings.loc[top.index]
    denom = np.sum(np.abs(top.values))
    if denom == 0:
        return np.nan
    return float(np.dot(top.values, ratings.values) / denom)
```

```python
def recommend_user_based(user_id: int, user_item: pd.DataFrame, user_sim: pd.DataFrame, k: int = 20, n: int = 10):
    candidates = get_unrated_items(user_item, user_id)
    if not candidates:
        return []

    preds = []
    for item_id in candidates:
        pred = predict_user_based(user_id, item_id, user_item, user_sim, k=k)
        if not np.isnan(pred):
            preds.append((item_id, pred))

    preds.sort(key=lambda x: x[1], reverse=True)
    return preds[:n]
```

```python
def recommend_item_based(user_id: int, user_item: pd.DataFrame, item_sim: pd.DataFrame, k: int = 20, n: int = 10):
    candidates = get_unrated_items(user_item, user_id)
    if not candidates:
        return []

    preds = []
    for item_id in candidates:
        pred = predict_item_based(user_id, item_id, user_item, item_sim, k=k)
        if not np.isnan(pred):
            preds.append((item_id, pred))

    preds.sort(key=lambda x: x[1], reverse=True)
    return preds[:n]
```

```python
some_user = user_item.index[0]

print("User-based recommendations:")
print(recommend_user_based(some_user, user_item, user_sim, k=30, n=10))

print("\nItem-based recommendations:")
print(recommend_item_based(some_user, user_item, item_sim, k=30, n=10))
```

```python
from sklearn.model_selection import train_test_split

train_df, test_df = train_test_split(df_cf, test_size=0.2, random_state=42)

# rebuild user_item theo train
train_ui = train_df.pivot_table(index='customer_id', columns='product_id', values='rating', aggfunc='mean')
train_ui_filled = train_ui.fillna(0)

# recompute sims
user_sim_train = pd.DataFrame(cosine_similarity(train_ui_filled), index=train_ui.index, columns=train_ui.index)
item_sim_train = pd.DataFrame(cosine_similarity(train_ui_filled.T), index=train_ui.columns, columns=train_ui.columns)

def rmse_on_test(test_df: pd.DataFrame, mode: str = "item", k: int = 20):
    se = []
    for u, i, r in test_df[['customer_id','product_id','rating']].itertuples(index=False):
        if mode == "user":
            pred = predict_user_based(u, i, train_ui, user_sim_train, k=k)
        else:
            pred = predict_item_based(u, i, train_ui, item_sim_train, k=k)

        if not np.isnan(pred):
            se.append((pred - r) ** 2)

    return float(np.sqrt(np.mean(se))) if se else np.nan

print("RMSE item-based:", rmse_on_test(test_df, mode="item", k=30))
print("RMSE user-based:", rmse_on_test(test_df, mode="user", k=30))
```

```python
def precision_at_k(predict_ids, test_ids, k):
    predict_k = predict_ids[:k]
    hit = len(set(predict_k) & set(test_ids))
    return hit / k
```

```python
def recall_at_k(predict_ids, test_ids, k):
    predict_k = predict_ids[:k]
    hit = len(set(predict_k) & set(test_ids))
    return hit / len(test_ids) if len(test_ids) > 0 else 0.0
```

```python
def f1_at_k(predict_ids, test_ids, k):
    p = precision_at_k(predict_ids, test_ids, k)
    r = recall_at_k(predict_ids, test_ids, k)
    return 2*p*r/(p+r) if (p+r) > 0 else 0.0
```

```python
def mrr(predict_ids, test_ids):
    for i, item in enumerate(predict_ids, start=1):
        if item in test_ids:
            return 1 / i
    return 0.0
```

```python
def ndcg(predict_ids, test_ids, k):
    dcg = 0.0
    for i, item in enumerate(predict_ids[:k], start=1):
        if item in test_ids:
            dcg += 1 / np.log2(i + 1)

    idcg = sum(1 / np.log2(i + 1) for i in range(1, min(len(test_ids), k) + 1))
    return dcg / idcg if idcg > 0 else 0.0
```

```python
def evaluate(test_ids, predict_ids, k):
    return {
        "Precision@K": precision_at_k(predict_ids, test_ids, k),
        "Recall@K": recall_at_k(predict_ids, test_ids, k),
        "F1@K": f1_at_k(predict_ids, test_ids, k),
        "MRR": mrr(predict_ids, test_ids),
        "NDCG": ndcg(predict_ids, test_ids, k)
    }
```

```python
def recommend_item_based_ids(
    user_id,
    train_ui,
    item_sim,
    k_neighbors=30,
    top_n=10
):
    if user_id not in train_ui.index:
        return []

    user_ratings = train_ui.loc[user_id]
    rated_items = user_ratings.dropna().index
    candidate_items = train_ui.columns.difference(rated_items)

    scores = []

    for item_id in candidate_items:
        pred = predict_item_based(
            user_id,
            item_id,
            train_ui,
            item_sim,
            k=k_neighbors
        )
        if not np.isnan(pred):
            scores.append((item_id, pred))

    scores.sort(key=lambda x: x[1], reverse=True)
    return [item for item, _ in scores[:top_n]]
```

```python
# map user -> list item trong test
test_items_by_user = (
    test_df
    .groupby('customer_id')['product_id']
    .apply(list)
    .to_dict()
)
```

```python
def evaluate_all_users(
    train_ui,
    item_sim,
    test_items_by_user,
    k_neighbors=30,
    top_k=10
):
    all_metrics = []

    for user_id, true_items in test_items_by_user.items():
        # bỏ user không có trong train
        if user_id not in train_ui.index:
            continue

        predicted_items = recommend_item_based_ids(
            user_id,
            train_ui,
            item_sim,
            k_neighbors=k_neighbors,
            top_n=top_k
        )

        if len(predicted_items) == 0:
            continue

        metrics = evaluate(
            test_ids=true_items,
            predict_ids=predicted_items,
            k=top_k
        )
        all_metrics.append(metrics)

    return pd.DataFrame(all_metrics)
```

```python
df_metrics = evaluate_all_users(
    train_ui=train_ui,
    item_sim=item_sim_train,
    test_items_by_user=test_items_by_user,
    k_neighbors=30,
    top_k=10
)

df_metrics.mean()
```

```python
for k in [5, 10, 15, 20, 25]:
    df_k = evaluate_all_users(
        train_ui=train_ui,
        item_sim=item_sim_train,
        test_items_by_user=test_items_by_user,
        k_neighbors=30,
        top_k=k
    )
    print(f"\nTop-{k}")
    print(df_k.mean())
```
