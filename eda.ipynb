```python
Q1 = df_products["price"].quantile(0.25)
Q3 = df_products["price"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df_products[
    (df_products["price"] < lower_bound) |
    (df_products["price"] > upper_bound)
]

len(outliers_iqr)
```

```python
df_reviews
```

```python
df_reviews.info()
```

```python
df_reviews = df_reviews.dropna(subset=["customer_id", "customer_name","title"], axis=0)
```

```python
df_reviews.isnull().sum()
```

```python
df_reviews["content"] = df_reviews["content"].fillna("")
```

```python
df_reviews.isnull().sum()
```

```python
df_reviews.info()
```

```python
sns.histplot(df_reviews["rating"], bins=5, discrete=True)
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")
plt.title("Distribution of Product Ratings")
plt.show()
```

```python
rating_pct = df_reviews["rating"].value_counts(normalize=True).sort_index() * 100
rating_pct
```

```python
rating_pct.plot(kind="bar")
plt.ylabel("Percentage (%)")
plt.title("Rating Distribution (%)")
plt.xticks(rotation=0)
plt.show()
```

```python
df_reviews["content"].head()
```

```python
df_reviews["content"].dtype
```

```python
df_reviews["has_content"] = df_reviews["content"].str.strip() != ""
```

```python
df_reviews["has_content"].value_counts(normalize=True) * 100
```

```python
df_reviews["content_len"] = df_reviews["content"].str.len()
df_reviews["content_len"].describe()
```

```python
sns.histplot(df_reviews[df_reviews["content_len"] > 0]["content_len"], bins=30)
plt.xlabel("Number of characters")
plt.title("Distribution of Review Content Length")
plt.show()
```

```python
p99 = df_reviews["content_len"].quantile(0.99)

sns.histplot(
    df_reviews[
        (df_reviews["content_len"] > 0) &
        (df_reviews["content_len"] <= p99)
    ]["content_len"],
    bins=30
)
plt.xlabel("Number of characters")
plt.title("Review Content Length (<= 99th percentile)")
plt.show()
```

```python
sns.boxplot(
    x=df_reviews[df_reviews["content_len"] > 0]["content_len"]
)
plt.title("Boxplot of Review Content Length")
plt.show()
```

```python
(df_reviews["content_len"] < 10).mean()
```

```python
df_reviews["word_count"] = df_reviews["content"].str.split().str.len()
df_reviews["word_count"].describe()
```

```python
sns.histplot(
    df_reviews[df_reviews["word_count"] > 0]["word_count"],
    bins=30
)
plt.xlabel("Number of words")
plt.title("Distribution of Review Word Count")
plt.show()
```

```python
sns.boxplot(
    x="rating",
    y="content_len",
    data=df_reviews[df_reviews["content_len"] > 0]
)
plt.title("Review Length by Rating")
plt.show()
```

```python
print("===== BASIC STATISTICS =====")
print("Total interactions:", len(df_reviews))
print("Total users:", df_reviews['customer_id'].nunique())
print("Total items:", df_reviews['product_id'].nunique())

print("\nAverage interactions per user:",
      df_reviews.groupby('customer_id').size().mean())

print("Average interactions per item:",
      df_reviews.groupby('product_id').size().mean())
```

```python
user_interactions = df_reviews.groupby('customer_id').size()

print("\n===== USER INTERACTION DISTRIBUTION =====")
print(user_interactions.describe())

print("\nUsers with < 3 interactions:",
      (user_interactions < 3).mean())
```

```python
user_interactions = df_reviews.groupby("customer_id").size()
item_interactions = df_reviews.groupby("product_id").size()
```

```python
plt.figure()
plt.hist(user_interactions, bins=50)
plt.xlabel("Number of interactions per user")
plt.ylabel("Number of users")
plt.title("User Interaction Distribution")
plt.show()
```

```python
plt.figure()
plt.hist(item_interactions, bins=50)
plt.xlabel("Number of interactions per item")
plt.ylabel("Number of items")
plt.title("Item Interaction Distribution")
plt.show()
```
