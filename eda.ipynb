```python
review_per_product = df_reviews.groupby("product_id")["review_id"].count()
print("Số review trung bình / sản phẩm:", review_per_product.mean())

plt.figure(figsize=(6,4))
sns.histplot(review_per_product, bins=50)
plt.title("Review count per product")
plt.xlabel("Số review")
plt.show()
```

```python
review_per_user = df_reviews.groupby("customer_id")["review_id"].count()
print("Số review trung bình / user:", review_per_user.mean())

plt.figure(figsize=(6,4))
sns.histplot(review_per_user, bins=50)
plt.title("Review count per user")
plt.xlabel("Số review")
plt.show()
```
