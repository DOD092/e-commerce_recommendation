```python
import re
import unicodedata
from underthesea import word_tokenize

def preprocess_short_description(text):
    if not isinstance(text, str):
        return ""

    # 1. Chuẩn hóa Unicode (tránh lỗi font, dấu)
    text = unicodedata.normalize("NFC", text)

    # 2. Chuyển về chữ thường
    text = text.lower()

    # 3. Loại URL (nếu có)
    text = re.sub(r"http\S+|www\S+", " ", text)

    # 4. Giữ chữ cái tiếng Việt + số
    text = re.sub(
        r"[^a-z0-9àáạảãâầấậẩẫăằắặẳẵ"
        r"èéẹẻẽêềếệểễ"
        r"ìíịỉĩ"
        r"òóọỏõôồốộổỗơờớợởỡ"
        r"ùúụủũưừứựửữ"
        r"ỳýỵỷỹđ\s]",
        " ",
        text
    )

    # 5. Chuẩn hóa khoảng trắng
    text = re.sub(r"\s+", " ", text).strip()

    # 6. Tách từ tiếng Việt (word segmentation)
    text = word_tokenize(text, format="text")

    # 7. Làm sạch lần cuối
    text = re.sub(r"\s+", " ", text).strip()

    return text
```

```python
df_products["short_description_clean"] = (
    df_products["short_description"]
    .apply(preprocess_short_description)
)
```
